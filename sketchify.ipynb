```python
def dodge(front,back):
    result=front*255/(255-back) 
    result[result>255]=255
    result[back==255]=255
    return result.astype('uint8')
```

```python
import numpy as np
def grayscale(rgb):
    return np.dot(rgb[...,:4], [0.299, 0.587, 0.114])
```

```python
#img ="http://static.cricinfo.com/db/PICTURES/CMS/263600/263697.20.jpg"
#img="C://Users//Chaitanya//Pictures//Sunny_Photos//Sunny1.jpeg"
img="C://Users//Chaitanya//Pictures//Sunny_Photos//WhatsApp Image 2020-05-26 at 1.31.15 PM (1).jpeg"
```

```python
import imageio
s = imageio.imread(img)
g=grayscale(s)
i = 255-g
import scipy.ndimage
b = scipy.ndimage.filters.gaussian_filter(i,sigma=10)
r= dodge(b,g)
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
plt.imshow(r, cmap="gray")
```
